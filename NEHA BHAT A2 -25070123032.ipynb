##DATA BINNING ND DAT FORMATIN USING PYTHON

```python
import pandas as pd
import numpy as np


data={
    'Product':['laptop','mobile','tablet','headphones','smartwatch','camera'],
    'Price':[55000,50000,45000,6000,7000,20000],
    'Units_sold':[20,30,40,50,60,18]

}
df=pd.DataFrame(data)
print("data")
print(df)
```

```python
# DATA BINNING
bins1 =[0,1000,30000,60000]
labels1 =['low','medium','high']
df['Price_category'] =pd.cut(df['Price'],bins=bins1,labels =labels1)
print(df)
```

```python

# binning units sold
bins2=[0,10,30,100]
labels2=['less','medium','more']
df['sales category'] = pd.cut(df['Units_sold'],bins=bins2,labels =labels2)
print(df)
```

```python
# data formatting
# convert data type
print(df.dtypes)
df['Units_sold']= df['Units_sold'].astype(float)
print(df.dtypes)
```

```python
# convert text to uppercase
df['Product']=df['Product'].str.upper()
print(df)
```

```python
# Rounding values
df['Price']=df['Price'].round(2)
print(df)
```

```python

# additional operations
sorted_df=df.sort_values(by='Price')
print(sorted_df)
```

```python
# display unique values
print(df['Price_category'].unique())
```
